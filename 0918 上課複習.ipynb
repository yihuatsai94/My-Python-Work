## while迴圈

條件成立時就會執行，不成立時就不會執行

```python
s = 0
while s<10:
    print(s)
    s = s + 1
#s+1的目的是為了不讓他繼續無限迴圈下去，因為是<10，所以到9之後就不會再執行下去了
```

```python
msg = "多拉A夢"

while (msg != "bye"):
    msg = input(">> ")
    print("拍拍!")
```
