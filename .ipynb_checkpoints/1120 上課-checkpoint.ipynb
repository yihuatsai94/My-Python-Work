```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## 模擬版資料做 SVM

```python
x = np.array([[6, 6], [10, 8], [3, 7], [8, 7]])
# 只能兩組數字是因為這樣才能畫在圖上
```

```python
x
```

```python
x[:, 0]
```

```python
plt.scatter(x[:, 0], x[:,1])
```

```python
y = np.array([1, 2, 1, 2])
```

```python
plt.scatter(x[:,0], x[:,1], c=y, cmap="Paired", s=150)
```

### Step 1. 打開函數學習機

```python
from sklearn.svm import SVC
```

```python
clf = SVC(gamma = 'auto')
# 開空白的學習機
```

### Step 2. 學習

```python
clf.fit(x, y)
```

解釋 warning

gamma 是在 svm 裡要讀的一個程式，負責控制分類的精確度

### Step 3. 預測

```python
y_predict = clf.predict(x)
```

```python
y_predict
```

### 鳶尾花數據庫

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
# 把資料讀進來
```

```python
print(iris.DESCR)
```

```python
x = iris.data
y = iris.target
```

```python
x[:5]
```

```python
y[:5]
# 第幾種鳶尾花
```

```python
X = x[:, :2]
```

```python
Y = y
# 用我們選出來的
```

```python
len(X)
```

### 區分訓練資料、測試資料

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
plt.scatter(X[:,0], X[:,1], c=Y, cmap='Paired')
# 鳶尾花數據庫
```

### Step 1. 開一台函數學習機

```python
clf = SVC(gamma='auto')
```

### Step 2. 訓練

```python
clf.fit(x_train, y_train)
```

### Step 3. 預測

```python
y_predict = clf.predict(x_test)
```

```python
y_predict
```

```python
y_test
```

```python
y_predict - y_test
```

```python
plt.scatter(x_test[:,0], x_test[:,1], c=y_predict - y_test)
```

```python
x0 = np.arange(3.8, 8.2, 0.02)
y0 = np.arange(1.8, 4.7, 0.02)

xm, ym = np.meshgrid(x0, y0)
P = np.c_[xm.ravel(), ym.ravel()]
z = clf.predict(P)

Z = z.reshape(xm.shape)
plt.contour(xm, ym, Z, cmap='Paired', alpha=0.3)
```
