```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.datasets import load_iris
from sklearn.svm import SVC
```

```python
iris = load_iris()
```

```python
x = iris.data
y = iris.target
```

#### 選擇 sepal length 跟 petal width 做預測

```python
X = x[:,[0,3]]
```

```python
Y = y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.27, random_state = 0)
```

```python
plt.scatter(X[:,0], X[:,1], c=Y, cmap="rainbow")
```

```python
clf = SVC(gamma='auto')
```

```python
clf.fit(x_train, y_train)
```

```python
y_predict = clf.predict(x_test)
```

```python
y_predict
```

```python
y_test
```

```python
y_predict - y_test
```

```python
plt.scatter(x_test[:,0], x_test[:,1], c=y_predict - y_test)
```

```python
x0 = np.arange(3.8, 7.6, 0.02)
y0 = np.arange(0.1, 2.6, 0.02)

xm, ym = np.meshgrid(x0, y0)
P = np.c_[xm.ravel(), ym.ravel()]
z = clf.predict(P)

Z = z.reshape(xm.shape)
plt.contour(xm, ym, Z, cmap='rainbow', alpha=0.2)

plt.scatter(x_test[:,0], x_test[:,1], cmap='rainbow', c=y_test)
```

由此可看出，使用 sepal length 跟 petal width 完全正確
