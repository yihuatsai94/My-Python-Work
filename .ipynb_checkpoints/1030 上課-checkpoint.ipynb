## 挑出我們要的資料

```python
dfs = df[:5]
```

```python
dfs
```

```python
df["加權"] = df.數學 * 2 + df.自然 * 3
```

```python
df.head()
```

```python
df.loc[:,"國文":"自然"]
#loc把想要要求的資料都求出來，全部都要
```

```python
df["總分"] = df.loc[:,"國文":"自然"].sum(axis=1)
```

```python
df.head()
```

```python
df.sort_values(by="總分")
```

```python
df.sort_values(by="總分", ascending=False).head()
#總分由低到高
```

```python
df.sort_values(by=["加權","總分"], ascending=False)
```

```python
df_sorted = df.sort_values(by=["加權", "總分"],ascending=False)
```

```python
df_sorted.head()
```

```python
df_sorted.index = range(1, 101)
```

```python
df_sorted.head()
```

```python
df_sorted.數學 == 15
```

```python
df_sorted[df_sorted.數學 == 15]
```

```python
df_sorted[(df_sorted.數學 == 15) & (df_sorted.英文 == 15)]
```

```python
df_sorted[(df_sorted.數學 == 15) | (df_sorted.英文 == 15)]
# | 為 or 的意思
```
