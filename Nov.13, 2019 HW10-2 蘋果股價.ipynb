```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pandas_datareader.data as web
```

```python
df = web.DataReader("AAPL", "yahoo", start="2014-9-1", end="2019-8-31")
```

```python
df.head()
```

### 看‘收盤價’

```python
P = df["Adj Close"]
```

```python
P.plot()
```

### 計算報酬率
$$\frac{P_t - P_{t-1}}{P_{t-1}}$$

```python
P.diff()
# 跟前一期的差別
```

```python
r = P.diff()/P
# 看百分比
```

```python
r.plot()
```

### 從150天前到現在報酬率的變化

```python
r[-150:].plot()
```

由此圖可以看出它的漲跌幅度很厲害

### 移動平均

使用移動平均可以讓線變得較平滑，較容易看出他的走向

```python
P.rolling(window=20).mean()
```

```python
P.rolling(window=20).mean().plot()
```

由此圖可看出在 2015 年初到年中股價有稍微上漲的趨勢，但在年中後稍微的跌宕起伏，到了 2017 年後有開始逐漸上漲的趨勢，但在 2019 年股價突然大跌。
