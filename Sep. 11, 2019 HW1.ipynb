```python
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()
#定義figure
ax = fig.gca(projection='3d')
#將此figure變成3D圖
X = np.arange(-4, 4, 0.15)
#x的範圍
Y = np.arange(-4, 4, 0.15)
#y的範圍
X, Y = np.meshgrid(X, Y)
#x, y平面網格
R = np.sqrt(X**2 + Y**2)
#x, y軸長度
Z = np.sin(R)
#z軸的高度
ax.plot_surface(X, Y, Z, rstride=3, cstride=3, cmap='coolwarm')
#3D曲面（橫軸、縱軸長跟寬為3, 色調為coolwarm）
plt.xkcd()
#Q版畫風
plt.show()
#顯示出此圖
```
