```python
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()
#figure
ax = fig.gca(projection='3d')
#figure3D
X = np.arange(-4, 4, 0.15)
#x
Y = np.arange(-4, 4, 0.15)
#y
X, Y = np.meshgrid(X, Y)
#x, y
R = np.sqrt(X**2 + Y**2)
#x, y
Z = np.sin(R)
#z
ax.plot_surface(X, Y, Z, rstride=3, cstride=3, cmap='coolwarm')
#3D3, coolwarm
plt.xkcd()
#Q
plt.show()
#
```
